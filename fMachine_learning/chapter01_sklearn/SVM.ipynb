```python
import pandas as pd
from sklearn import svm,metrics
```

```python
csv = pd.read_csv('bmi.csv')
csv
```

```python
data_df = pd.DataFrame(csv)
data_df.columns

data_learn = data_df[['height',' weight']]
data_result = data_df[' label']
#print(data_result)
clf = svm.SVC(gamma='scale')
clf.fit(data_learn,data_result)

pre1 = clf.predict(data_learn)

#학습하고 정확도를 구하세요.

ac_score = metrics.accuracy_score(data_result,pre1)
print(ac_score)


test = [[160,64],[180,64],[170,64]]

test_df = pd.DataFrame(test)

pre = clf.predict(test_df)
print(pre)

```
