```python
import mglearn
import pandas as pd 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc


font_location = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

data = pd.read_csv("./data/order1.csv")
#print(data.iloc[:,0:])

kmeans = KMeans(n_clusters = 5)
kmeans.fit(data.iloc[:,0:])

print("클러스터 레이블 ", kmeans.labels_) 


mglearn.discrete_scatter(data.iloc[:,0], data.iloc[:,1], kmeans.labels_)
plt.legend(["클러스터 0", "클러스터 1", "클러스터 2","클러스터 3","클러스터 4"], loc='best')
plt.xlabel("ticket ")
plt.ylabel("shopping ")
plt.show()

print(kmeans.predict([[50,10]]))
```
