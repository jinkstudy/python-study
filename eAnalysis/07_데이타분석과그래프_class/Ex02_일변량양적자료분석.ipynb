  ( 데이터가 가지는 값의 종류가 많기 때문에 )
  
** 양적자료로 빈도와 백분율을 구하려면 구간별 새로운 자료를 생성하고 구함

### 양적자료 시각화 

1. 히스토그램

2. 박스플롯


```python
# 데이타 읽어오기

import pandas as pd

# 인덱스가 2번 생성되기에 기존 인덱스를 인덱스로 지정
diamonds = pd.read_csv('data/diamonds.csv', index_col='Unnamed: 0')
diamonds.head()
```

```python
#Price에 대해서 히스토그램
import matplotlib.pyplot as plt
import seaborn as sns

axes = plt.subplot()
axes = sns.distplot(diamonds['price']);
#axes.set_title('Total Bill histogram');


```

```python
#price에 대해서 박스플롯
plt.subplot()
axes = sns.boxplot(diamonds['price']);
```

### [양적자료를 질적자료로 새로 생성 ]

```python
diamonds['gprice'] = pd.cut(diamonds['price'],[0,5000,10000,15000,20000], labels=['5000','10000','15000','20000'])
diamonds.head(20)
diamonds[diamonds['gprice'] == '10000']
```

```python
#gprice  - > 막대그래프

a = diamonds.groupby('gprice').count()['price']
plt.subplot()
axes = sns.barplot(x=a.index,y=a.values) # 자동으로 평균을 구해서 그래프 그려줌
```
