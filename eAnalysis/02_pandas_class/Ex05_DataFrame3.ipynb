## 데이타 필터링

분석에 필요한 데이타를 거르는 작업 ( 데이타 추출 )

1. 열단위 추출 : df['컬럼명']

2. 행단위 추출 : df.loc[인덱스명칭] / df.iloc[인덱스위치]

    - loc[] : 명칭 기반 인덱싱
    - iloc[] : 위치 기반 인덱싱
    
    현재 인덱스가 글자로 지정되지 않았기에 loc[]/iloc[] 모두 가능
    

        - loc과 iloc 차이
            - df.loc[-1]  # 에러
            - df.iloc[-1] # 마지막번호

```python
# 데이타 추출

import pandas as pd

df = pd.read_excel('http://qrc.depaul.edu/excel_files/presidents.xlsx')
df

# 1. 열단위 추출
#  대통령 이름 / 취임나이 / 출신주

df[['President','Age at inauguration','State elected from']]
```

```python
# 2. 행단위 추출
# 조지워싱턴 / 링컨 / 오바마 대통령
""" 
    loc[] : 명칭 기반 인덱싱
    iloc[] : 위치 기반 인덱싱
    
    현재 인덱스가 글자로 지정되지 않았기에 loc[]/iloc[] 모두 가능
"""
# (1) loc

df.loc[[0,13,34]]

df.loc[(df['President'] == 'George Washington')| (df['President'] == 'Abraham Lincoln') | (df['President'] == 'Barack Obama')]

df2 = df.set_index('President')
df2.
```

```python
# (2) iloc
df.iloc[[0,13,34]]
```

마지막행의 데이타 추출

```python
df.iloc[-1]
```
