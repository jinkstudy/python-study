 ## 미국 대통령 신장(키) 데이타를 분석 
 
 
        ` 전체적인 기본 통계량
        ` 평균신장
        ` 그래프로 시각화

```python
import pandas as pd

data = pd.read_csv('data/president_heights.csv')
data
```

```python
# 기본 통계랑 (평균 / 표준편차 /최솟값 / 최댓값 /중앙값)

print(data.mean())
print("*"*50)
print(data.median())
print("*"*50)
print(data.std())
print("*"*50)
print(data.min())
print("*"*50)
print(data.max())
```
