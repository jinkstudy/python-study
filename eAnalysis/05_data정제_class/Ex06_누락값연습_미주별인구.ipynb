## 1. 데이타를 로딩 후 데이타 정보 확인

```python
# 1. 데이타를 로딩 후 데이타 정보 확인
import pandas as pd

pop = pd.read_csv('./data/state-population.csv')    # 주별 인구
areas = pd.read_csv('./data/state-areas.csv')       # 주의 면적
abbrevs = pd.read_csv('./data/state-abbrevs.csv')   # 주이름 약자

# 데이타 살펴보기
print(pop)

print()
print(areas)

# 행수 확인
#pop.shape

#abbrevs에 Puerto Rico PR 추가
abbrevs.loc[51]=['Puerto Rico','PR']
print(abbrevs)
```

---
## 2. 데이타 연결
        
    1- 인구 데이타(pop)와 주이름 약어 데이타(abbrevs)를 연결
            ( 각 주의 전체 이름과 인구를 볼 수 있도록 )
    2- 각 주의 인구 데이타(pop)와 면적 데이타(areas)를 연결
            ( 각 주의 면적당 인구를 계산하여 인구 밀도를 구하기 )



```python
result1= pop.merge(abbrevs,left_on = 'state/region',right_on = 'abbreviation',how ='outer')
result1

result2= result1.merge(areas,left_on = 'state',right_on = 'state', how ='outer')
result2

#2010년만 sorting
result_2010 = result2[(result2['year'] == 2010)]
result_2010




```
