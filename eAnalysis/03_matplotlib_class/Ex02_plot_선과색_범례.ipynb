```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
x = np.linspace(0, 10, 100)  # 0~10 사이의 수를 100개 뽑기
plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-0), color='blue')  # 색상 이름
plt.plot(np.sin(x-1), color='g')  #색상 코드
plt.plot(np.sin(x-2), color='#FFDD33')  # 16진수
plt.plot(np.sin(x-3), color='chartreuse')  # html 색상이름
plt.plot(np.sin(x-4), color='0.7')  # 0 ~ 1 회색 색
plt.plot(np.sin(x-5), color=(1.0, 0.2, 0.5))  # rgb 튜플 형식


```

```python
""" 2. 선스타일 """
# plt.plot(np.sin(x-0), linestyle='solid')
# plt.plot(np.sin(x-0), linestyle='dashed')
# plt.plot(np.sin(x-0), linestyle='dotted')
plt.plot(np.sin(x-0), linestyle='dashdot')

# plt.plot(np.sin(x-0), linestyle=':')
# plt.plot(np.sin(x-0), linestyle='-.')
# plt.plot(np.sin(x-0), linestyle='--')
# plt.plot(np.sin(x-0), linestyle='-')


```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """

plt.plot(np.sin(x-0), 'g-o', label='first')
plt.plot(np.sin(x-1), 'c--p', label='second')
plt.plot(np.sin(x-2), '-.k', label='third')

plt.legend(loc=1)  # 우상단
plt.legend(loc=4)  # 우하단
```
